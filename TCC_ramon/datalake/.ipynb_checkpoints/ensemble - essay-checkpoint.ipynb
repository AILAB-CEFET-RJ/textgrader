```python
from dags import config
import os
import pandas as pd
import glob

i = 1

super_lista = []

for i in range(1,9):
    path = os.path.join(config.ESSAY_CONTAINER,'predictions','test',f'set_{i}','domain_1','*')
    file_list = glob.glob(path)

    listao = []

    for file in file_list:
        df = pd.read_parquet(file)
        df = df[['text_id','text_set','prediction','score']]
        file_name = file.split("\\")[-1].split('.')[0]
        df['file_name'] = file_name
        listao.append(df)

    dfzao = pd.concat(listao)
    super_lista.append(dfzao)
    

big_df = pd.concat(super_lista)
big_df['erro'] = big_df['prediction'] - big_df['score']
```

```python
 
# Priorização para o cálculo do preço referência
def seleciona(row):
    if row['Contagem'] == row['Unicos']:
        return row['Mediana']
    elif row['N_modas'] > 1:
        return row['Mediana']
    else:
        return row['Moda']

moda_mediana['moda_previsao'] = moda_mediana.apply(seleciona, axis = 1)

moda_mediana
```
