```python
from dags import config
import os
import pandas as pd
import glob
from sklearn.metrics import cohen_kappa_score
import numpy as np

i = 1

super_lista = []

for i in range(1,9):
    path = os.path.join(config.ESSAY_CONTAINER,'predictions','test',f'set_{i}','domain_1','*')
    file_list = glob.glob(path)

    listao = []

    for file in file_list:
        df = pd.read_parquet(file)
        df = df[['text_id','text_set','prediction','score']]
        file_name = file.split("\\")[-1].split('.')[0]
        df['file_name'] = file_name
        listao.append(df)

    dfzao = pd.concat(listao)
    super_lista.append(dfzao)
    

big_df = pd.concat(super_lista)
big_df['erro'] = big_df['prediction'] - big_df['score']
```

```python
moda_df = big_df.groupby(['text_set', 'text_id'])['prediction'].agg(Contagem ='count', Unicos = 'nunique', Mediana = 'median',Media = 'mean', Moda = lambda x: pd.Series.mode(x)[0], N_modas = lambda x: len(pd.Series.mode(x))).reset_index()

# Priorização para o cálculo do preço referência
def seleciona(row):
    if row['Contagem'] == row['Unicos']:
        return row['Mediana']
    elif row['N_modas'] > 1:
        return row['Mediana']
    else:
        return row['Moda']

moda_df['moda_previsao'] = moda_df.apply(seleciona, axis = 1)

score = big_df.groupby(['text_set', 'text_id'])['score'].mean().reset_index()

moda_df['score'] = score['score']
moda_df = moda_df.astype(int)
```

```python
moda_df['difference'] = moda_df['moda_previsao'] - moda_df['score']
moda_df['difference'] = np.where(moda_df['difference'] < -4, 'menor que -4', np.where(moda_df['difference'] > 4, 'maior que 4', moda_df['difference']))
moda_df
```

```python
diferencas = moda_df.pivot_table(index = 'text_set',values= 'text_id', columns = 'difference',aggfunc = 'count').replace(np.NaN,0).reset_index()
#diferencas.columns = ['Conjunto','0','1','2','3']
diferencas = diferencas[['text_set','menor que -4','-4','-3','-2','-1','0','1','2','3','4','maior que 4']]
diferencas
```

```python
cohen = moda_df.groupby(['text_set']).apply(lambda x:cohen_kappa_score(x['moda_previsao'],x['score'],weights = 'quadratic'))
cohen = cohen.reset_index()
cohen.columns = ['Conjunto','kappa de cohen']
cohen
```
