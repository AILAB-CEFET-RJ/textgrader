```python
from nltk import word_tokenize
```

```python
def word_count(x):
    lista = word_tokenize(x)
    return len(lista)
```

```python
def word_count_unique(x):
    lista = word_tokenize(x)
    return len(lista.unique())
```

```python
import pandas as pd
df = pd.read_parquet('corrected_texts.parquet')

df['Essay Text'].apply(lambda x: word_count(x))
```
