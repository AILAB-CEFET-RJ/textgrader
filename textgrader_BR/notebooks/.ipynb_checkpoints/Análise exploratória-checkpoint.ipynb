```python
%reload_kedro

import pandas as pd
import numpy as np 
from textgrader.pipelines.config import *
import seaborn as sns
import matplotlib.pyplot as plt
```

# Análise exploratória

## preprocessamento dos dados

```python
df_train = catalog.load('basic_train')
df_test = catalog.load('basic_test')

df = pd.concat([df_train,df_test])
df = df.rename(columns = TARGETS_1_RENAME)
df1 = df[df['conjunto'] == 1]
 
```

## Análise da correlação entre os targets

```python
df1[['c1','c2','c3','c4','c5']].corr()
```

## Palavras, palavras únicas e sentenças

### palavras x palavras únicas

```python
sns.scatterplot(df1, x ='word_count',y = 'unique_word_count')
```

### palavras x sentenças 

```python
sns.scatterplot(df1, x ='word_count',y = 'sentence_count')
```

### palavras únicas x sentenças

```python
sns.scatterplot(df1, x ='unique_word_count',y = 'sentence_count')
```

## Contagem de redaçoes por score (mostrada em percentuais) populaçao cerca de 7500 redações

```python
test = df1[['c1','c2','c3','c4','c5']]
test = test.melt()
test = test.reset_index()
test =  test.groupby(['variable','value']).count().reset_index()
test = test.set_index(['value','variable']).unstack('variable').reset_index()

test.columns = ['nota','c1','c2','c3','c4','c5']
test = test.fillna(0)

for i in ['c1','c2','c3','c4','c5']:
    test[i] = test[i]/sum(test[i]) * 100
test
```
